1. Difference between np.array, np.asarray, and np.asanyarray
All three functions are used to create NumPy arrays, but they differ in how they handle the input data and the types of arrays they produce.

np.array:

Creates a new array and copies the data by default.
It always creates a new array object, even if the input is already a NumPy array, unless you explicitly specify copy=False.
Recommended when you need a guarantee that the resulting object is a base-level ndarray.
np.asarray:

Converts the input to a NumPy array, but does not copy the data if the input is already a NumPy array (i.e., if the input is already an array, it returns the same object).
Useful when you want to convert inputs like lists or tuples to an array without duplicating memory if it's unnecessary.
np.asanyarray:

Similar to np.asarray, but it preserves the input type if the input is a subclass of ndarray. For example, if you pass a matrix object, it returns a matrix instead of converting it to a base ndarray.
Use it if you want to retain subclassed array types like np.matrix or other custom types derived from ndarray.

2. Difference between Deep Copy and Shallow Copy
When copying an object in Python (e.g., lists or arrays), there are two main types of copies:

Shallow Copy:

A shallow copy creates a new object, but does not recursively copy nested objects within the original. Instead, it inserts references to the original nested objects.
Modifying the elements of the shallow copy affects the original nested elements (and vice versa).
Deep Copy:

A deep copy creates a new object and recursively copies all objects found in the original. This means that changes to the deep copy do not affect the original object, and vice versa.
The nested elements are fully copied, resulting in a complete separation from the original object.
Key Differences Summary:
Shallow Copy: Copies the top-level structure only; nested objects are still shared between the original and the copy.
Deep Copy: Recursively copies all elements and nested objects; no references are shared.
Use Case:

Use shallow copy for simple structures when you don't need to modify nested objects.
Use deep copy when you need a completely independent copy of the entire structure, including all nested objects.

Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:
 right y-axis for 'Humidity').
 b) Label the x-axis as 'Date'.
 a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
c) Set the title of the plot as 'Temperature and Humidity Over Time'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset in a Pandas DataFrame
# Generate a date range
dates = pd.date_range(start='2024-01-01', periods=30)

# Create random data for Temperature and Humidity
np.random.seed(42)  # For reproducibility
temperature = np.random.uniform(low=20, high=35, size=30)  # Temperatures between 20°C and 35°C
humidity = np.random.uniform(low=30, high=90, size=30)  # Humidity percentage between 30% and 90%

# Create the DataFrame
data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Create the plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Temperature on the left y-axis
ax1.plot(data['Date'], data['Temperature'], color='tab:red', label='Temperature (°C)')
ax1.set_xlabel('Date')  # Label for the x-axis
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Step 3: Create the second y-axis for Humidity
ax2 = ax1.twinx()  # Create a twin Axes sharing the x-axis
ax2.plot(data['Date'], data['Humidity'], color='tab:blue', label='Humidity (%)')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Step 4: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Step 5: Display the plot
fig.tight_layout()  # Adjust layout to accommodate y-axis labels
plt.show()
```

 Q15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:
 a) Plot a histogram of the data with 30 bins.
 b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Create a NumPy array containing 1000 samples from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.randn(1000)  # Generates 1000 samples from a standard normal distribution (mean=0, std=1)

# Step 2: Plot a histogram of the data with 30 bins
count, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram')

# Step 3: Overlay a line plot representing the PDF of the normal distribution
# Calculate the PDF values for the corresponding bins
pdf = stats.norm.pdf(bins, np.mean(data), np.std(data))
plt.plot(bins, pdf, color='red', linewidth=2, label='PDF')

# Step 4: Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Step 6: Display the legend
plt.legend()

# Step 7: Show the plot
plt.show()
```

Q17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create two random arrays (x and y) with values ranging from -10 to 10
np.random.seed(42)  # For reproducibility
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

# Step 2: Create a DataFrame to store x, y, and their corresponding quadrant
data = pd.DataFrame({'x': x, 'y': y})

# Step 3: Define a function to classify points into quadrants based on their position relative to the origin
def get_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant I'  # Top-right
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant II'  # Top-left
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'  # Bottom-left
    else:
        return 'Quadrant IV'  # Bottom-right

# Step 4: Apply the function to classify each point
data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Step 5: Create the Seaborn scatter plot, coloring points by quadrant
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='x', y='y', hue='Quadrant', palette='Set2', data=data)

# Step 6: Add labels to the axes and set the title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Step 7: Add a legend
plt.legend(title='Quadrants')

# Step 8: Display the plot
plt.axhline(0, color='black', linestyle='--', linewidth=1)  # Horizontal line at y=0
plt.axvline(0, color='black', linestyle='--', linewidth=1)  # Vertical line at x=0
plt.show()
```

 Q18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Prepare the data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2π
y = np.sin(x)  # Sine wave values

# Step 2: Create a Bokeh figure
output_notebook()  # Use this line if you are running in Jupyter Notebook
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Step 3: Add the line to the figure
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Step 4: Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Step 5: Show the plot
show(p)
```

Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.models import HoverTool

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 30, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values})

# Create a Bokeh figure
p = figure(title='Random Categorical Bar Chart', x_axis_label='Category', y_axis_label='Value')

# Create a bar chart
p.vbar(x=data['Category'], top=data['Value'], width=0.5, color='color', legend='Category')

# Add color based on values
color_mapper = p.color_mapper(field_name='Value', palette='Viridis256')
p.add_tools(HoverTool(tooltips=[('Category', '@Category'), ('Value', '@Value')]))

# Output the plot
output_file('categorical_bar_chart.html')
show(p)
```
